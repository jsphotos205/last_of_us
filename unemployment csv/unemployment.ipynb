```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/Users/johnchristophersloan/the_last_of_us/unemployment csv/output.csv')

df[df.eq('Kentucky').any(1)]

kentucky = df[(df.values.ravel() == 'Kentucky').reshape(df.shape).any(1)]

wolfe = kentucky[(kentucky.values.ravel() == 'Wolfe County').reshape(kentucky.shape).any(1)]
powell = kentucky[(kentucky.values.ravel() == 'Powell County').reshape(kentucky.shape).any(1)]
lee = kentucky[(kentucky.values.ravel() == 'Lee County').reshape(kentucky.shape).any(1)]



year = kentucky['Year']
wolfe_year = wolfe['Year']
powell_year = powell['Year']
lee_year = lee['Year']

wolfe_month = wolfe['Month']
powell_month = powell['Month']
lee_month = lee['Month']

wolfe_unemployment_rate = wolfe['Rate']
powell_unemployment_rate = powell['Rate']
lee_unemployment_rate = lee['Rate']
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(x=wolfe_year, y = wolfe_unemployment_rate, data= wolfe_unemployment_rate, label='Wolfe County')
sns.lineplot(x=powell_year, y = powell_unemployment_rate, data=powell_unemployment_rate, label='Powell County')
sns.lineplot(x=lee_year, y=lee_unemployment_rate, data=lee_unemployment_rate, label='Lee County')
plt.legend(loc='best')
#how do you add a legend?
```
