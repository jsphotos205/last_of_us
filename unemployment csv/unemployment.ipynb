```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme()
```

```python
#import the data from a csv file found on kaggle of unemployment rates by counties throughout the USA
df = pd.read_csv('/Users/johnchristophersloan/the_last_of_us/unemployment csv/output.csv')

#from that csv file isolate Kentucky's information
df[df.eq('Kentucky').any(1)]
kentucky = df[(df.values.ravel() == 'Kentucky').reshape(df.shape).any(1)]

#wanting to focus on just the tri county area around the Red River Gorge
wolfe = kentucky[(kentucky.values.ravel() == 'Wolfe County').reshape(kentucky.shape).any(1)]
powell = kentucky[(kentucky.values.ravel() == 'Powell County').reshape(kentucky.shape).any(1)]
lee = kentucky[(kentucky.values.ravel() == 'Lee County').reshape(kentucky.shape).any(1)]

wolfe_year = wolfe['Year']
powell_year = powell['Year']
lee_year = lee['Year']

wolfe_month = wolfe['Month']
powell_month = powell['Month']
lee_month = lee['Month']

wolfe_unemployment_rate = wolfe['Rate']
powell_unemployment_rate = powell['Rate']
lee_unemployment_rate = lee['Rate']
```

```python
#create linegraph that shows unemployment rates in tri county area
#want to look at contributing factors for unemployment from other data sources as well as contributing factors that increases employment in said areas
plt.figure(figsize=(12,6))

sns.lineplot(x=wolfe_year, y = wolfe_unemployment_rate, data= wolfe_unemployment_rate, label='Wolfe County')
sns.lineplot(x=powell_year, y = powell_unemployment_rate, data=powell_unemployment_rate, label='Powell County')
sns.lineplot(x=lee_year, y=lee_unemployment_rate, data=lee_unemployment_rate, label='Lee County')
plt.title('Red River Gorge Tri-County Unemployment Rates by Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=wolfe_year, y=wolfe_unemployment_rate, data=wolfe_unemployment_rate)
plt.xticks(rotation=45)
plt.title('Wolfe County Unemployment Rates by Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=powell_year, y=powell_unemployment_rate, data=powell_unemployment_rate)
plt.xticks(rotation=45)
plt.title('Powell County Unemployment Rates by Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=lee_year, y=lee_unemployment_rate, data=lee_unemployment_rate)
plt.xticks(rotation=45)
plt.title('Lee County Unemployment Rates by Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```
