```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme()
```

```python
#import the data from a csv file found on kaggle of unemployment rates by counties throughout the USA
df = pd.read_csv('/Users/johnchristophersloan/the_last_of_us/unemployment csv/output.csv')

#from that csv file isolate Kentucky's information
df[df.eq('Kentucky').any(1)]
kentucky = df[(df.values.ravel() == 'Kentucky').reshape(df.shape).any(1)]

#wanting to focus on just the tri county area around the Red River Gorge
wolfe = kentucky[(kentucky.values.ravel() == 'Wolfe County').reshape(kentucky.shape).any(1)]
powell = kentucky[(kentucky.values.ravel() == 'Powell County').reshape(kentucky.shape).any(1)]
lee = kentucky[(kentucky.values.ravel() == 'Lee County').reshape(kentucky.shape).any(1)]
breathitt = kentucky[(kentucky.values.ravel() == 'Breathitt County').reshape(kentucky.shape).any(1)]
owsley = kentucky[(kentucky.values.ravel() == 'Owsley County').reshape(kentucky.shape).any(1)]

#as a control we could use Fayette county (lexington) and Jefferson county (louisville)
fayette = kentucky[(kentucky.values.ravel() == 'Fayette County').reshape(kentucky.shape).any(1)]
jefferson = kentucky[(kentucky.values.ravel() == 'Jefferson County').reshape(kentucky.shape).any(1)]

wolfe_year = wolfe['Year']
powell_year = powell['Year']
lee_year = lee['Year']
breathitt_year = breathitt['Year']
owsley_year = owsley['Year']

fayette_year =fayette['Year']
jefferson_year =jefferson['Year']

wolfe_month = wolfe['Month']
powell_month = powell['Month']
lee_month = lee['Month']
breathitt_month = breathitt['Month']
owsley_month = owsley['Month']

fayette_month = fayette['Month']
jefferson_month = jefferson['Month']

wolfe_unemployment_rate = wolfe['Rate']
powell_unemployment_rate = powell['Rate']
lee_unemployment_rate = lee['Rate']
breathitt_unemployment_rate = breathitt['Rate']
owsley_unemployment_rate = owsley['Rate']

fayette_unemployment_rate = fayette['Rate']
jefferson_unemployment_rate = jefferson['Rate']
```

```python
#create linegraph that shows unemployment rates in counties surrounding the red river gorge area
#want to look at contributing factors for unemployment from other data sources as well as contributing factors that increases employment in said areas
#use of jefferson and fayette as realtion to large kentucky city

plt.figure(figsize=(12,6))

sns.lineplot(x=wolfe_year, y = wolfe_unemployment_rate, data= wolfe_unemployment_rate, label='Wolfe County')
sns.lineplot(x=powell_year, y = powell_unemployment_rate, data=powell_unemployment_rate, label='Powell County')
sns.lineplot(x=lee_year, y=lee_unemployment_rate, data=lee_unemployment_rate, label='Lee County')
sns.lineplot(x=breathitt_year, y=breathitt_unemployment_rate, data=breathitt_unemployment_rate, label='Breathitt County')
sns.lineplot(x=owsley_year, y=owsley_unemployment_rate, data=owsley_unemployment_rate, label='Owsley County')
sns.lineplot(x=fayette_year, y=fayette_unemployment_rate, data=fayette_unemployment_rate, label='Fayette County')
sns.lineplot(x=jefferson_year, y=jefferson_unemployment_rate, data=jefferson_unemployment_rate, label='Jefferson County')

plt.title('Unemployment Rates by Year in Relation to Fayette and Jefferson County')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.legend(loc='best')
plt.show()
```
