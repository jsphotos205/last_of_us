```python
import requests
from pprint import pprint as pp
import pandas as pd
```

```python
# https://cryptingup.com/apidoc/#introduction
response = requests.get('https://cryptingup.com/api/assets/?ID=btc')
main_info = response.json()
```

```python
crytocurrency_df = pd.DataFrame(columns=['coin','price','one hour change','one day change', 'one week change', 'update'])
crytocurrency_df
```

```python
first_main_info = main_info['assets'][0]

first_id = main_info['assets'][0]['asset_id']
first_price = main_info['assets'][0]['price']
first_change_1h = main_info['assets'][0]['change_1h']
first_change_24h = main_info['assets'][0]['change_24h']
first_change_7d = main_info['assets'][0]['change_7d']

first_created_at = main_info['assets'][0]['created_at']
first_description = main_info['assets'][0]['description']
first_updated_at = main_info['assets'][0]['updated_at']
```

```python
def get_first_currency(main_info):
    first_main_info = main_info['assets'][0]
    first_id = main_info['assets'][0]['asset_id']
    first_price = main_info['assets'][0]['price']
    first_change_1h = main_info['assets'][0]['change_1h']
    first_change_24h = main_info['assets'][0]['change_24h']
    first_change_7d = main_info['assets'][0]['change_7d']
    first_created_at = main_info['assets'][0]['created_at']
    first_description = main_info['assets'][0]['description']
    first_updated_at = main_info['assets'][0]['updated_at']

    return first_main_info, first_id, first_price, first_change_1h, first_change_24h, first_change_7d, first_updated_at
```

```python
get_first_currency(main_info)
```

```python
class First:

    for coin in first_main_info:
        print(first_id)
        break
    for price in first_main_info:
        print(first_price)
        break
    for one_hour_change in first_main_info:
        print(first_change_1h)
        break
    for one_day_change in first_main_info:
        print(first_change_24h)
        break
    for one_week_change in first_main_info:
        print(first_change_7d)
        break
    for update in first_main_info:
        print(first_updated_at)
        break
```

```python
crytocurrency_df = crytocurrency_df.append({'coin': first_id,
                                            'price': first_price,
                                            'one hour change': first_change_1h,
                                            'one day change': first_change_24h,
                                            'one week change': first_change_7d,
                                            'update':first_updated_at}, ignore_index=True)
crytocurrency_df
```
